1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution 
of age with respect to each gender along with the information about whether they survived 
or not. (Column names : 'sex' and 'age')
2. Write observations on the inference from the above statistics.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
df
```

```python
df['Age'] = df['Age'].fillna(np.mean(df['Age']))
df['Age']
df['Cabin'] = df['Cabin'].fillna(df['Cabin'].mode()[0])
df['Cabin']
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
male_df = df[df['Sex'] == "male"]
```

```python
female_df = df[df['Sex'] == 'female']
```

```python
male_survived = male_df[male_df['Survived'] == 1]['Age']
male_not_survived = male_df[male_df['Survived'] == 0]['Age']

female_survived = female_df[female_df['Survived'] == 1]['Age']
female_not_survived = female_df[female_df['Survived'] == 0]['Age']
```

```python
import matplotlib.pyplot as plt

data = [male_survived, male_not_survived, female_survived, female_not_survived]

fig, ax = plt.subplots()

boxes = ax.boxplot(data, labels=['Male Survived', 'Male Not Survived', 'Female Survived', 'Female Not Survived'])

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']
ax.set_title('Age Distribution by Gender and Survival Status')
ax.set_xlabel('Gender and Survival Status')
ax.set_ylabel('Age')
for i, box in enumerate(boxes['boxes']):
    box.set_markerfacecolor(colors[i])

plt.show()
```

```python
import seaborn as sns

data = pd.DataFrame({'Gender': df['Sex'], 'Age': df['Age'], 'Survived': df['Survived']})

# Set style and context for seaborn
sns.set_style('whitegrid')
sns.set_context('talk')

# Create box plot
ax = sns.boxplot(x='Gender', y='Age', hue='Survived', data=data, palette=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'], showfliers=False)

# Set title and axis labels
ax.set_title('Age Distribution by Gender and Survival Status')
ax.set_xlabel('Gender')
ax.set_ylabel('Age')

# Show the plot
plt.show()
```
